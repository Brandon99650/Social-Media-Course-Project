```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_91 = pd.read_csv(
    os.path.join("data","fit","91", "paras.csv")
)
df_91[df_91["title"].isin(["2022蔣萬安","2022陳時中"])].to_csv(
    "mr.csv"
)
```

```python
plt.figure(dpi = 600)
j = np.load(os.path.join("data","fit","91","2022蔣萬安.npy"))
clock = np.load(os.path.join("data","fit","91","2022陳時中.npy"))
plt.plot(
    list(i for i in range(j.shape[0])), j, label="Mr.J - fitting"
)
plt.plot(
    list(i for i in range(clock.shape[0])), clock, label="Mr.Clock - fitting"
)
plt.legend()
plt.savefig("TaipeiMy.jpg")
plt.show()
```

```python
df_president=pd.read_csv(
    os.path.join("data","fit","president","paras.csv")
)
df_president.head()
```

```python
plt.figure(dpi = 600)

z = np.load(os.path.join("data","fit","president","宋楚瑜.npy"))
tee = np.load(os.path.join("data","fit","president","蔡英文.npy"))
k_finsh = np.load(os.path.join("data","fit","president","韓國瑜.npy"))
plt.plot(
    list(i for i in range(tee.shape[0])), tee, label="TEE - fitting"
)
plt.plot(
    list(i for i in range(k_finsh.shape[0])), k_finsh, label="K_fish - fitting"
)
plt.plot(
    list(i for i in range(z.shape[0])), z, label="Yu - fitting"
)
plt.legend()
plt.savefig("president2020.jpg")
plt.show()
```
